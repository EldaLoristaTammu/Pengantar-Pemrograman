```python
import pandas as pd #mengolah data dalam bentuk tabel
from scipy.stats.mstats import gmean, hmean #menghitung rata-rata geometrik dan harmonik
import matplotlib.pyplot as plt #untuk membuat histogram dan menampilkan garis vertikal untuk rata-rata

my_dataset = pd.read_excel('Smith_glass_post_NYT_data.xlsx', sheet_name='Supp_traces') #memuat dataset dari file Excel

a_mean = my_dataset.Dy.mean() #menghitung rata-rata aritmatika dari kolom 'Dy'
g_mean = gmean(my_dataset['Dy']) #menghitung rata-rata geometrik dari kolom 'Dy' menggunakan fungsi gmean dari scipy.stats.mstats
h_mean = hmean(my_dataset['Dy']) #menghitung rata-rata harmonik dari kolom 'Dy' menggunakan fungsi hmean dari scipy.stats.mstats

print ('-------') #mencetak garis pemisah
print ('arithmetic mean') #mencetak label untuk rata-rata aritmatika
print ("{0:.1f} [ppm]".format(a_mean)) #mencetak nilai rata-rata aritmatika dengan format satu angka desimal diikuti oleh satuan ppm
print ('-------') #mencetak garis pemisah

print ('geometric mean') #mencetak label untuk rata-rata geometrik
print ("{0:.1f} [ppm]".format(g_mean)) #mencetak nilai rata-rata geometrik dengan format satu angka desimal diikuti oleh satuan ppm
print ('-------') #mencetak garis pemisah

print ('harmonic mean') #mencetak label untuk rata-rata harmonik
print ("{0:.1f} [ppm]".format(h_mean)) #mencetak nilai rata-rata harmonik dengan format satu angka desimal diikuti oleh satuan ppm
print ('-------') #mencetak garis pemisah

fig, ax = plt.subplots() #membuat sebuah figure dan axes untuk plotting
ax.hist(my_dataset.Dy, bins='auto', density=True, edgecolor='k', label='Measurements Hist', alpha=0.8, color='pink') #membuat histogram dari kolom 'Dy' dengan jumlah bin otomatis, menampilkan probabilitas density, memberikan warna pink pada histogram, dan menambahkan label untuk legenda
ax.axvline(a_mean, color='green', label='Arithmetic mean', linewidth=2) #menambahkan garis vertikal pada posisi rata-rata aritmatika dengan warna hijau, label untuk legenda, dan ketebalan garis 2
ax.axvline(g_mean, color='blue',  label='Geometric mean', linewidth=2) #menambahkan garis vertikal pada posisi rata-rata geometrik dengan warna biru, label untuk legenda, dan ketebalan garis 2
ax.axvline(h_mean, color='red',  label='Harmonic mean', linewidth=2) #menambahkan garis vertikal pada posisi rata-rata harmonik dengan warna merah, label untuk legenda, dan ketebalan garis 2
ax.set_xlabel('Dy [ppm]') #memberikan label pada sumbu x dengan teks 'Dy [ppm]'
ax.set_ylabel('Probability density') #memberikan label pada sumbu y dengan teks 'Probability density'
ax.legend() #menampilkan legenda pada plot
```

```python
import pandas as pd #mengolah data dalam bentuk tabel
import numpy as np #untuk menghitung histogram dan mencari nilai bin dengan frekuensi tertinggi
import matplotlib.pyplot as plt #untuk membuat histogram dan menampilkan garis vertikal untuk nilai modal

my_dataset = pd.read_excel('Smith_glass_post_NYT_data.xlsx', sheet_name='Supp_traces') #memuat dataset dari file Excel
hist, bin_edges = np.histogram(my_dataset['Ba'], bins= 20, density=True) #menghitung histogram dari kolom 'Ba' dengan jumlah bin 20 dan menampilkan probabilitas density, hasilnya disimpan dalam variabel hist dan bin_edges
modal_value = (bin_edges[hist.argmax()] +  bin_edges[hist.argmax()+1])/2 #menghitung nilai modal dengan mencari bin yang memiliki frekuensi tertinggi menggunakan argmax() pada histogram, kemudian mengambil rata-rata dari batas bawah dan batas atas bin tersebut untuk mendapatkan nilai modal

print ('modal value: {0:.0f} [ppm]'.format(modal_value)) #mencetak nilai modal dengan format tanpa angka desimal diikuti oleh satuan ppm

fig, ax = plt.subplots() #membuat sebuah figure dan axes untuk plotting
ax.hist(my_dataset.Ba, bins=20, density=True, edgecolor='k', label="Measurements Hist", alpha=0.8, color= 'green' ) #membuat histogram dari kolom 'Ba' dengan jumlah bin 20, menampilkan probabilitas density, memberikan warna default pada histogram, dan menambahkan label untuk legenda
ax.axvline(modal_value, color="orange", label="Modal value", linewidth=2) #menambahkan garis vertikal pada posisi nilai modal dengan warna oranye, label untuk legenda, dan ketebalan garis 2
ax.set_xlabel('Ba [ppm]') #memberikan label pada sumbu x dengan teks 'Ba [ppm]'
ax.set_ylabel('Probability density') #memberikan label pada sumbu y dengan teks 'Probability density'
ax.legend() #menampilkan legenda pada plot

'''
Output: modal value: 277 [ppm] #mencetak nilai modal yang telah dihitung sebelumnya dengan format tanpa angka desimal diikuti oleh satuan ppm
'''
```

```python
import pandas as pd #mengolah data dalam bentuk tabel
import matplotlib.pyplot as plt #untuk membuat histogram dan menampilkan garis vertikal untuk nilai varians dan standar deviasi

my_dataset = pd.read_excel('Smith_glass_post_NYT_data.xlsx', sheet_name='Supp_traces') #memuat dataset dari file Excel

variance =  my_dataset['La'].var() #menghitung nilai varians dari kolom 'La' menggunakan metode var() dari pandas, hasilnya disimpan dalam variabel variance
stddev =  my_dataset['La'].std() #menghitung nilai standar deviasi dari kolom 'La' menggunakan metode std() dari pandas, hasilnya disimpan dalam variabel stddev

print ('-------') #mencetak garis pemisah
print ('Variance') #mencetak label untuk varians
print("{0:.0f} [square ppm]".format(variance)) #mencetak nilai varians dengan format tanpa angka desimal diikuti oleh satuan square ppm
print ('-------') #mencetak garis pemisah
print ('Standard Deviation') #mencetak label untuk standar deviasi
print("{0:.0f} [ppm]".format(stddev)) #mencetak nilai standar deviasi dengan format tanpa angka desimal diikuti oleh satuan ppm
print ('-------') #mencetak garis pemisah

fig, ax = plt.subplots() #membuat sebuah figure dan axes untuk plotting
ax.hist(my_dataset.La, bins= 20, density = True, edgecolor='y', label='Measurements Hist', color='lightcoral') #membuat histogram dari kolom 'La' dengan jumlah bin 20, menampilkan probabilitas density, memberikan warna default pada histogram, dan menambahkan label untuk legenda
ax.axvline(my_dataset['La'].mean()-stddev, color='purple', label=r'mean - 1$\sigma$', linewidth=2) #menambahkan garis vertikal pada posisi mean - 1*stddev dengan warna ungu, label untuk legenda, dan ketebalan garis 2
ax.axvline(my_dataset['La'].mean()+stddev, color='green', label=r'mean + 1$\sigma$', linewidth=2) #menambahkan garis vertikal pada posisi mean + 1*stddev dengan warna hijau, label untuk legenda, dan ketebalan garis 2
ax.axvspan(my_dataset['La'].mean()-stddev, my_dataset['La'].mean()+stddev, alpha=0.1, color='orange', label=r'mean $\pm$ 1$\sigma$') #menambahkan area berwarna oranye antara mean - 1*stddev dan mean + 1*stddev dengan transparansi 0.1 dan label untuk legenda
ax.set_xlabel('La [ppm]') #memberikan label pada sumbu x dengan teks 'La [ppm]'
ax.set_ylabel('Probability density') #memberikan label pada sumbu y dengan teks 'Probability density'
ax.legend() #menampilkan legenda pada plot

'''
Output:
-------
Variance
14021 [square ppm]
-------
Standard Deviation
118 [ppm]
-------
'''
```

```python
import pandas as pd #mengolah data dalam bentuk tabel
import numpy as np #untuk menghitung persentil dan interkuartil range (IQR)
import matplotlib.pyplot as plt #untuk membuat histogram dan menampilkan garis vertikal untuk nilai kuartil pertama (Q1), kuartil ketiga (Q3), dan area interkuartil range (IQR)

my_dataset = pd.read_excel('Smith_glass_post_NYT_data.xlsx', sheet_name='Supp_traces') #memuat dataset dari file Excel

q1 = np.percentile(my_dataset.Ho, 25, method='midpoint') #menghitung nilai kuartil pertama (Q1) dari kolom 'Ho' menggunakan fungsi percentile dari numpy dengan metode 'midpoint' untuk menangani data yang memiliki jumlah elemen genap
q3 = np.percentile(my_dataset.Ho, 75, method='midpoint') #menghitung nilai kuartil ketiga (Q3) dari kolom 'Ho' menggunakan fungsi percentile dari numpy dengan metode 'midpoint' untuk menangani data yang memiliki jumlah elemen genap
  
iqr = q3 - q1 # Interquaritle range (IQR) #menghitung interkuartil range (IQR) dengan mengurangi nilai kuartil pertama (Q1) dari nilai kuartil ketiga (Q3)

print ('-------') #mencetak garis pemisah
print ('Interquaritle range (IQR): {0:.0f} [ppm]'.format(iqr)) #mencetak nilai interkuartil range (IQR) dengan format tanpa angka desimal diikuti oleh satuan ppm
print ('-------') #mencetak garis pemisah

fig, ax = plt.subplots() #membuat sebuah figure dan axes untuk plotting
ax.hist(my_dataset.Ho, bins='auto', density=True, edgecolor='y', label='Measurements Hist', color='#FF0000', linewidth=1) #membuat histogram dari kolom 'Ho' dengan jumlah bin otomatis, menampilkan probabilitas density, memberikan warna merah pada histogram, dan menambahkan label untuk legenda
ax.axvline(q1, color='pink', label='Q1', linewidth=2) #menambahkan garis vertikal pada posisi kuartil pertama (Q1) dengan warna pink, label untuk legenda, dan ketebalan garis 2
ax.axvline(q3, color='yellow', label='Q3', linewidth=2) #menambahkan garis vertikal pada posisi kuartil ketiga (Q3) dengan warna kuning, label untuk legenda, dan ketebalan garis 2
ax.axvspan(q1, q3, alpha=0.1, color='green', label='Interquaritle range (IQR)') #menambahkan area berwarna hijau antara kuartil pertama (Q1) dan kuartil ketiga (Q3) dengan transparansi 0.1 dan label untuk legenda
ax.set_xlabel('Ho [ppm]') #memberikan label pada sumbu x dengan teks 'Ho [ppm]'
ax.set_ylabel('Probability density') #memberikan label pada sumbu y dengan teks 'Probability density'
ax.legend() #menampilkan legenda pada plot

'''
Output:
-------
Interquaritle range (IQR): 164 [ppm]
-------
'''
```

```python
import pandas as pd #mengolah data dalam bentuk tabel
import numpy as np #untuk menghitung rata-rata, median, dan histogram untuk mencari nilai modal
import matplotlib.pyplot as plt #untuk membuat histogram dan menampilkan garis vertikal untuk nilai rata-rata, median, dan modal value

my_dataset = pd.read_excel('Smith_glass_post_NYT_data.xlsx', sheet_name='Supp_traces') #memuat dataset dari file Excel

a_mean = my_dataset.Er.mean() #menghitung rata-rata aritmatika dari kolom 'Er' menggunakan metode mean() dari pandas, hasilnya disimpan dalam variabel a_mean

median = my_dataset.Er.median() #menghitung nilai median dari kolom 'Er' menggunakan metode median() dari pandas, hasilnya disimpan dalam variabel median

hist, bin_edges = np.histogram(my_dataset['Er'], bins=20, density=True) #menghitung histogram dari kolom 'Er' dengan jumlah bin 20 dan menampilkan probabilitas density, hasilnya disimpan dalam variabel hist dan bin_edges
modal_value = (bin_edges[hist.argmax()] +  bin_edges[hist.argmax()+1])/2 #menghitung nilai modal dengan mencari bin yang memiliki frekuensi tertinggi menggunakan argmax() pada histogram, kemudian mengambil rata-rata dari batas bawah dan batas atas bin tersebut untuk mendapatkan nilai modal

fig, ax = plt.subplots() #membuat sebuah figure dan axes untuk plotting
ax.hist(my_dataset.Er, bins=20, density=True, edgecolor='r', label="Measurements Hist", color= '#FFFF00' , linewidth='1') #membuat histogram dari kolom 'Er' dengan jumlah bin 20, menampilkan probabilitas density, memberikan warna kuning pada histogram, dan menambahkan label untuk legenda
ax.axvline(modal_value, color='red', label='Modal Value', linewidth=2) #menambahkan garis vertikal pada posisi nilai modal dengan warna merah, label untuk legenda, dan ketebalan garis 2
ax.axvline(median, color='blue',  label='Median Value', linewidth=2) #menambahkan garis vertikal pada posisi nilai median dengan warna biru, label untuk legenda, dan ketebalan garis 2
ax.axvline(a_mean, color='pink',  label='Arithmetic mean', linewidth=2) #menambahkan garis vertikal pada posisi nilai rata-rata dengan warna pink, label untuk legenda, dan ketebalan garis 2
ax.set_xlabel('Er [ppm]') #memberikan label pada sumbu x dengan teks 'Er [ppm]'
ax.set_ylabel('Probability density') #memberikan label pada sumbu y dengan teks 'Probability density'
ax.legend() #menampilkan legenda pada plot

```

```python
import numpy as np #untuk menghitung rata-rata, median, standar deviasi, dan histogram untuk mencari nilai modal
from scipy.stats import skew #untuk menghitung koefisien skewness menggunakan fungsi skew dari scipy.stats

a_mean = my_dataset.Th.mean() #menghitung rata-rata aritmatika dari kolom 'Th'
median = my_dataset.Th.median() #menghitung nilai median dari kolom 'Th'
hist, bin_edges = np.histogram(my_dataset['Th'], bins=20, density=True) #menghitung histogram dari kolom 'Th' dengan jumlah bin 20 dan menampilkan probabilitas density
modal_value = (bin_edges[hist.argmax()] + bin_edges[hist.argmax()+1])/2 #menghitung nilai modal dengan mencari bin yang memiliki frekuensi tertinggi
standard_deviation = my_dataset['Th'].std() #menghitung standar deviasi dari kolom 'Th'

a1 = (a_mean - modal_value) / standard_deviation #menghitung Pearson's first coefficient of skewness dengan rumus (mean - mode) / standard deviation
a2 = 3 * (a_mean - median) / standard_deviation #menghitung Pearson's 2nd moment of skewness dengan rumus 3 * (mean - median) / standard deviation
g1 = skew(my_dataset['Th']) #menghitung Fisher-Pearson's coefficient of skewness menggunakan fungsi skew dari scipy.stats pada kolom 'Th'

print ('-------') #mencetak garis pemisah
print ("Pearson's first coefficient of skewness: {:.2f}".format(a1)) #mencetak nilai Pearson's first coefficient of skewness dengan format dua angka desimal
print ("Pearson's 2nd moment of skewness: {:.2f}".format(a2)) #mencetak nilai Pearson's 2nd moment of skewness dengan format dua angka desimal
print ("Fisher-Pearson's coefficient of skewness: {:.2f}".format(g1)) #mencetak nilai Fisher-Pearson's coefficient of skewness dengan format dua angka desimal
print ('-------') #mencetak garis pemisah

'''
Output: 
-------
Pearson's first coefficient of skewness: 0.74
Pearson's 2nd moment of skewness: 0.66
Fisher-Pearson's coefficient of skewness: 1.26
-------
'''
```
